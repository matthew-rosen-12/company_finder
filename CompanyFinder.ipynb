```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
import time

print('To use this program, you must have Google Chrome and ChromeDriver installed on your computer in the Downloads Folder.')
time.sleep(1)
print('Use this program to search for companies, e.g. \'Australian Trucking Companies\'.')
print('Once the program completes, a file including names of matching companies, each company\'s domain name, and 3 employees associated with each company will be downloaded.')
time.sleep(1)
# add prerequisites for running this program
time.sleep(1)
want_to_search = raw_input("Enter what you want to search: ") 
want_to_search_searchable = want_to_search.replace(' ', '%20')
time.sleep(1)
adapt_email = raw_input("What is your adapt.io email? Note: Your account must be through adapt.io, not Google.")
adapt_password = raw_input("What is your adapt.io password?")
time.sleep(1)
#ask user for filters
print("What number of employees would you like to filter for? Type in 0 to not filter for number of employees.")
print('Type in one of the following:')
print('0')
print('Freelancer')
print('2-9')
print('10-49')
print('50-249')
print('250-999')
print('1,000-9,999')
print('10,000+')
employees_size = raw_input()
while employees_size != '0' and employees_size != 'Freelancer' and employees_size != '2-9' and employees_size != '10-49' and employees_size != '50-249' and employees_size != '250-999' and employees_size != '1,000-9,999' and employees_size != '10,000+':
    print('That is not a valid input')
    print("What number of employees would you like to filter for? Type in 0 to not filter for number of employees.")
    print('Type in one of the following:')
    print('0')
    print('Freelancer')
    print('2-9')
    print('10-49')
    print('50-249')
    print('250-999')
    print('1,000-9,999')
    print('10,000+')
    employees_size = raw_input()
      
if employees_size == '0':
    employees_size = ''
elif employees_size == 'Freelancer':
    employees_size = 'Freelancer'
elif employees_size == '2-9':
    employees_size = '2+-+9'
elif employees_size == '10-49':
    employees_size = '10+-+49'
elif employees_size == '50-249':
    employees_size = '50+-+249'
elif employees_size == '250-999':
    employees_size = '250+-+999'
elif employees_size == '1,000-9,999':
    employees_size = '1%2C000+-+9%2C999'
elif employees_size == '10,000+':
    employees_size = '10%2C000%2B'
      
time.sleep(1)
print('Now obtaining information')

DRIVER_PATH = 'Downloads/chromedriver'
driver = webdriver.Chrome(executable_path=DRIVER_PATH)
driver.implicitly_wait(30);

driver.get('https://clutch.co/search/cse?query=' + want_to_search_searchable)

cookie = driver.find_element_by_xpath("//a[@id='CybotCookiebotDialogBodyButtonAccept']").click()
time.sleep(5)
link = driver.find_element_by_xpath("//a[@class='gs-title']")
link.click()
time.sleep(5)
driver.switch_to.window(driver.window_handles[1])
time.sleep(5)
company_names = []
company_websites = []

def get_companies():
    companies = driver.find_elements_by_xpath("//h3[@class='company-name']")
    websites = driver.find_elements_by_xpath("//span[text()='Visit Website']/parent::a")
    
    if not companies:
        return False
    
    for website in websites:
        company_websites.append(website.get_attribute('href'))

    for company in companies:
        company_names.append(company.text)

    return True;
        
base_url = driver.current_url

filtered_url = '&avg_hrly_rate=&employees=' + employees_size +'&client_focus=&industry_focus=&'

driver.get(base_url+filtered_url)

no = 0
keepGoing = True
while keepGoing:
    driver.get(base_url + '?page=' + str(no) + '&' + filtered_url)
    no = no + 1
    keepGoing = get_companies()
    
print(len(company_names), 'companies found')
time.sleep(1)
print(company_names)

driver.get('https://www.adapt.io/login.htm?slc=web&login=true')
email_field = driver.find_element_by_xpath("//input[@id='emailSignIn']")
email_field.send_keys('ym287@cornell.edu')
password_field = driver.find_element_by_xpath("//input[@name='passwordSignin']")
password_field.send_keys('713136Myc')
password_field.send_keys(Keys.ENTER)
driver.get('https://leads.adapt.io/advanced-search/contact#search')
executives_list = []
for domain in company_websites:
    search_bar= driver.find_element_by_xpath("//input[@type='text']")
    search_bar.send_keys(domain)
    search_bar.send_keys(Keys.ENTER)
    single_names = driver.find_elements_by_xpath("//div[@class='name contact-card-wrapper']")[:3]
    executives_names = []
    for element in single_names:
        executives_names.append(element.text)
    executives_list.append(executives_names)
    
import csv
with open(want_to_search+'.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow([want_to_search])
    writer.writerow(["Company Name", "Website"])
    writer.writerow(["Employees"])
    writer.writerow('')
    for i in range(len(company_names)):
        writer.writerow([company_names[i],company_websites[i]])
        writer.writerow(executives_list[i])
        writer.writerow('')
print()
print('Search has finished. A CSV file containing matching companies, company domains, and company employees has been created in local storage.')
```
